```python
from numpy import median
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("data01.csv")

total_house_value = 0
total_rows = len(df)

for x in df.index:
    total_house_value += df.loc[x, "median_house_value"]

mean_of_median = total_house_value / total_rows
median_of_house_value = df['median_house_value'].median()

print("-----------------House Value-----------------")
print("Median of total house value: ", median_of_house_value)
print("Mean of total house value: ", mean_of_median)
print("-----------------Rooms-----------------")

total_rooms_less_than_500 = 0
room_amount = 0

for x in df.index:
    room_amount += df.loc[x, "total_rooms"]

print("Total rooms: ",room_amount)
mean_rooms = room_amount / total_rows
median_rooms = df['total_rooms'].median()

for x in df.index:
   if df.loc[x, "total_rooms"] < 500:
    df.loc[x, "total_rooms"] = median_rooms
    total_rooms_less_than_500 += 1

print("Median of total rooms: ", median_rooms)
print("Mean of total rooms: ", mean_rooms)
print("Total rooms less than 500: ", total_rooms_less_than_500)
print("-----------------Housing Age-----------------")

total_housing_age = 0

print("-----------------Rows With A Median Housing Age Under 25-----------------")

for x in df.index:
  total_housing_age += df.loc[x, "housing_median_age"]
  if df.loc[x, "housing_median_age"] < 25:
    print(df.loc[x])

mean_housing_age = total_housing_age / total_rows
print("Mean housing age: ", mean_housing_age)

df.plot(kind = 'scatter', x = 'total_rooms', y = 'total_bedrooms')
plt.show()

df.plot(kind = 'pie', x = 'median_income', y = 'median_house_value')
plt.show()

```
